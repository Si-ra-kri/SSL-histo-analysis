```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt

DATA_PATH = r"C:\Users\siva\Documents\Winter Intern Project\data\classifier\train"

print("Data path:", DATA_PATH)
print("Number of files:", len(os.listdir(DATA_PATH)))
```

```python
files = [f for f in os.listdir(DATA_PATH) if f.endswith('.npy')]
print("Total .npy files:", len(files))

plt.figure(figsize=(12, 6))
for i, f in enumerate(random.sample(files, 10)):  # show 10 random images
    path = os.path.join(DATA_PATH, f)
    data = np.load(path, allow_pickle=True).item()  # each file is a dict {'image', 'mask'}
    img = data['image']  # only take the image part

    plt.subplot(2, 5, i + 1)
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(img.astype(np.uint8))
    plt.axis('off')
    plt.title(f.split('_')[0])

plt.tight_layout()
plt.show()
```

```python
file = random.choice(files)
data = np.load(os.path.join(DATA_PATH, file), allow_pickle=True).item()

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(data['image'].astype(np.uint8))
plt.title('Image'); plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(data['mask'], cmap='gray')
plt.title('Mask'); plt.axis('off')

plt.suptitle(file, fontsize=8)
plt.show()
```
